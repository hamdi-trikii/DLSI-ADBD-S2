```python
import numpy as np
from scipy import stats
tab=[124,101,115,126,114,112,138,85,138,96,130,116,132]

#1/calculer la moyenne le médiane et le mode
moyenne= np.mean(tab)
print("moyenne= ",moyenne.round(2))

median=np.median(tab)
print("median= ",median)

mode=stats.mode(tab)
print("mode= ",mode)


#2/calculer l'étendu ,la variance et lécart-type
etendue = np.ptp(tab)
print('etendue=' ,etendue)

variance= np.var(tab,ddof=1)
print("variance= ",variance.round(2))

cart_type = np.std(tab)
print("l'ecart type =",cart_type.round(2))

#3/
"""les données montre une dispression relativement large dépruis les voyageurs
nationaux lorss de leurs déplacements dans divers pays en 2016 (l'ecart-type >5)
avec une variavilité de 15.6 dollar entre le valeur le plus bas et le plus élevé"""

#4/
tab[-1]=175
print("***********************************")
moyenne= np.mean(tab)
print("moyenne= ",moyenne)

median=np.median(tab)
print("median= ",median)

mode=stats.mode(tab)
print("mode= ",mode)


#calculer l'étendu ,la variance et lécart-type

etendue = np.max(tab) - np.min(tab)
print('etendue=' ,etendue)

variance= np.var(tab)
print("variance= ",variance)

cart_type = np.std(tab)
print("l'ecart type =",cart_type)
```
